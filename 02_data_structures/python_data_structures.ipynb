# SET OPERATIONS

### UNION

- the union of sets can be performed using the union() method or the pipe (|) operator. Both methods combine all unique elements from the involved sets into a new set, discarding any duplicates. 

- It willmerge all elements

```python
s5 ={ 1,2,3,4,5}
s6 ={ 4,5,6,7,8}
s7 ={8,9,10}
```

```python
s5.union (s6)
```

```python
s5.union (s6,s7)
```

```python
s5|s6|s7               ### short cut of union is |
```

```python
s7|s6|s5
```

### INTERSECTION

- Intersection() method or the & operator. Both methods return a new set containing only the elements that are common

```python
s5 ={ 1,2,3,4,5}
s6 ={ 4,5,6,7,8}
s7 ={8,9,10}
```

```python
s5.intersection(s6)
```

```python
s6 & s7      ### short cut of intersection is &
```

```python
s5&(s7)
```

### DIFFERENCE   -  (-)

- The difference of two sets contains the elements that are present in the first set but not in the second set.
- In simple words: elements of set A that are not in set B

```python
s5 ={ 1,2,3,4,5}
s6 ={ 4,5,6,7,8}
s7 ={8,9,10}
```

```python
s5.difference (s6)
```

```python
s6-s5
```

```python
s5- s7
```

```python
s6-s7
```

### DIFFERENCE_UPDATE

- The difference_update() method removes elements found in another set from the original set.

```python
s5 ={ 1,2,3,4,5}
s6 ={ 4,5,6,7,8}
s7 ={8,9,10}
```

```python
s5.difference_update(s6)
```

```python
s5
```

```python
s6.difference_update(s7)
```

```python
s6
```

### SYMMETRIC_DIFFERNCE    - (^)
 
- (The symmetric difference of two sets is the set of elements that are present in either of the sets but not in both.)
- In simlple terms it will remove the common elements and print the output

```python
s5 ={ 1,2,3,4,5}
s6 ={ 4,5,6,7,8}
s7 ={8,9,10}
```

```python
s5^ s6
```

### SYMMETRIC_DIFFERENCE_UPDATE -  (^=)

- The symmetric_difference_update() method updates the original set by keeping only the elements that are present in either of the sets but not in both.
- In simple terms it modifies the set in place and does not return a new set.

```python
print(s5)
print(s6)
print(s7)
```

```python
s5 ={ 1,2,3,4,5}
s6 ={ 4,5,6,7,8}
s7 ={8,9,10}
```

```python
s5.symmetric_difference_update(s6)
```

```python
s5
```
