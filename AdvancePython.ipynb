## 2) While Loop - 

* Used when you don’t know how many times the loop should run, Loop runs as long as a condition is TRUE
  
* A while loop repeatedly executes a block of code as long as a given condition is true.

If the condition becomes false, the loop stops.


### Basic Syntax of while Loop



- while condition :
  
    statement(s)

### Flow of Execution

1️⃣ i = 1 → variable initialized

2️⃣ while i <= 5 → condition checked

3️⃣ If True → enter loop body

4️⃣ print(i) executes

5️⃣ i = i + 1 → value changes

6️⃣ Go back to step 2

7️⃣ When condition becomes False → loop stops

```python
i = 1
while i <= 5:
    print(i)
    i = i+1            ### Print → increment → recheck; runs while true, stops on false
```

## Break Statements 

- The break statement in Python brings control out of the loop

### Flow of Execution

1) Loop starts

2) Condition is checked

3) Loop body executes

4) When break is encountered:

5) Loop stops instantly

6) Remaining iterations are skipped

7) Program continues after the loop

### Continue Statements

- continue is used to skip the current iteration and move to the next iteration of the loop,  it give entire output

### Flow Execution

1) Loop starts

2) Condition is checked

3) Code runs until continue

4) When continue is hit:

5) Remaining statements in the loop are skipped

6) Control goes back to loop condition

7) Next iteration starts

### Example 1 

```python
for i in range(1,11):
    print(i)
```

```python
for i in range(1,11):
    if i == 3:  
        continue 
    print(i) 

## This loop runs from 1 to 10 and checks each value of i.
# When i becomes 3, continue skips printing it, so all numbers except 3 are printed
```
