# PYTHON NUMBER SYSTEM

 Python number system has been divided in to four types 
 
1) Binary number system(base 2) - 0 & 1(0b)
2) Octal number system (base 8) - 0,1,2,3,4,5,6,7 (0o)
3) Decimal number system (base 10) ------ 0x
4) Hexa decimal number system (base 16) ------ 0x ( a-10, b-11, c-12, d-14, e-14, f-15) ------(0-f)

### How to swap 2 variable in python 

```python
a = 5
b = 6
```

```python
temp = a
a = b
b = temp
```

```python
print(a)
print(b)
```

```python
a2 = 5
b2 = 6      ## Another method of swaping using arithmetic operation
```

```python
a2 = a2+b2
b2 = a2 - b2
a2 = a2 - b2 
```

```python
print(a2)
print(b2)
```

```python
a1 = 7
b1 = 8    ## Another method of swaping using equating both the values 
```

```python
a1,b1 = b1,a1 
```

```python
print(a1)
print(b1)
```
